```python
df["Pregnancies"].value_counts().plot(kind="bar",color = ["blue","skyblue"])
plt.title("Pregnancies Distribution")
plt.xlabel("Pregnancies")
plt.ylabel("count")
plt.show()
```

```python
plt.hist(df["Age"], bins= 10,color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.hist(df["Glucose"], bins= 10,color="pink")
plt.title(" DiGlucosestribution")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(x="SkinThickness",
                y= "Insulin",
                hue = "Pregnancies",data=df)
plt.title("SkinThickness VS Insulin")
plt.xlabel("SkinThickness")
plt.ylabel("Insulin")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Outcome', data=df)
plt.title("Count Plot of Diabetes Outcome")
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap of Diabetes Dataset")
plt.show()
```
